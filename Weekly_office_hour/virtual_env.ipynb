```bash

```

```bash
pip install package_name==1.0.0

To update a package to its latest version, you can use the --upgrade option:

```bash
pip install --upgrade package_name
```

To uninstall a package, you can use the uninstall command:

```bash

You can also list all the installed packages and their versions by using the list command:

```bash

What are the drawbacks of installing packages globally?
-------------------------------------------------------

While installing packages globally using pip or conda may seem convenient, it has some drawbacks that can cause problems in the long run. Some of these drawbacks are:

- You can only have one version of a package installed at a time for a given Python interpreter. This can cause issues when different projects require different versions of the same package or have incompatible dependencies.
  
- Installing packages globally may interfere with the system's Python environment and cause conflicts with other applications or libraries that rely on it.

- Installing packages globally may require administrative privileges or root access, which can pose security risks or limit your control over your system.

- Installing packages globally may make it harder to reproduce your project's environment on another machine or share it with others.

## Virtual Environments

> A Python virtual environment helps us create an isolated working copy of a software project that uses a specific version of Python interpreter together with specific versions of a number of external libraries installed into that virtual environment. Python virtual environments are implemented as directories with a particular structure within software projects, containing links to specified dependencies allowing isolation from other software projects on your machine that may require different versions of Python or external libraries.



## Managing Python Virtual Environments



There are several commonly used command line tools for managing Python virtual environments:


- [venv](https://docs.python.org/3/library/venv.html), available by default from the standard Python distribution from Python 3.3+

- [virtualenv](https://virtualenv.pypa.io/en/latest/user_guide.html), needs to be installed separately but supports both Python 2.7+ and Python 3.3+versions

- [pipenv](https://pypi.org/project/pipenv/), created to fix certain shortcomings of virtualenv

- [conda](https://docs.conda.io/en/latest/), package and environment management system (also included as part of the Anaconda Python distribution often used by the scientific community)

- [poetry](https://python-poetry.org), a modern Python packaging tool which handles virtual environments automatically

> While there are pros and cons for using each of the above, all will do the job of managing Python virtual environments for you and it may be a matter of personal preference which one you go for.

### Exporting and Sharing Environments

- Once you have created an environment and installed all the necessary packages, you may want to share this environment with your team or colleagues. 

- To do this, you can export the environment to a YAML file using the following command:

```bash
conda env export > environment.yml 
```

> This command will create a file named environment.yml that contains all the packages and dependencies installed in the environment. You can then share this file with your team or colleagues, who can use it to recreate the environment on their own machines.

### Recreate the environment 


```bash
conda env create -f environment.yml

```
