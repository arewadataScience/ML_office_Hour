- These labels are typically placed next to each slice and provide information about the category or value that the slice represents. 

- For example, in a pie chart showing the distribution of different fruit values in a DataFrame, the label slices might be the names of the different fruits, such as â€˜Appleâ€™, â€˜Bananaâ€™, and â€˜Orangeâ€™.

## Explode, shade, and rotate slices

In addition to the basic pie chart, this demo shows a few optional features:

- offsetting a slice using explode

- add a drop-shadow using shadow

- custom start angle using startangle

> The example below orders the slices, separates (explodes) them, and rotates them.



```python
explode = (0, 0.3, 0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

plt.pie(sizes, 
        explode=explode, 
        labels=labels, 
        autopct='%1.1f%%',
        shadow=True, 
        startangle=90)
plt.show()
```

> The default startangle is 0, which would start the first slice ("Frogs") on the positive x-axis. This example sets startangle = 90 such that all the slices are rotated counter-clockwise by 90 degrees, and the frog slice starts on the positive y-axis.

## Alternatives to Pie Chart

- The barplot is the best alternative to pie plots. 

```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]

df = pd.DataFrame({'labels': labels, 'sizes': sizes})
df
```

```python
# Create a figure and a subplot
fig, ax = plt.subplots()

# Create a bar chart of the data
ax.bar(df['labels'], df['sizes'])

# Set the axis labels
ax.set_xlabel('Labels')
ax.set_ylabel('Sizes')

# Display the chart
plt.show()
```

- If you have many values to display, you can also consider a lollipop plot that is a bit more elegant 

```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs', 'Fox', 'Lox', 'Cam', 'Ma', 'Lam', 'Fat']
sizes = [15, 30, 45, 10,8,43,56,71,37,68]

df = pd.DataFrame({'labels': labels, 'sizes': sizes})

# Create a figure and a subplot
fig, ax = plt.subplots()

# Create a scatter plot of the data
ax.scatter(df['sizes'], df.index, color='red')

# Draw vertical lines from the scatter points to the x-axis
ax.vlines(df['sizes'], ymin=df.index, ymax=-1, color='red')

# Set the y-axis ticks and labels
ax.set_yticks(df.index)
ax.set_yticklabels(df['labels'])

# Set the axis labels
ax.set_xlabel('Sizes')
ax.set_ylabel('Labels')

# Display the chart
plt.show()
```

> You can also do horizontal lollipop

```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs', 'Fox', 'Lox', 'Cam', 'Ma', 'Lam', 'Fat']
sizes = [15, 30, 45, 10, 8, 43, 56, 71, 37, 68]

df = pd.DataFrame({'labels': labels, 'sizes': sizes})

# Create a figure and a subplot
fig, ax = plt.subplots()

# Create a scatter plot of the data
ax.scatter(df.index, df['sizes'], color='red')

# Draw horizontal lines from the scatter points to the y-axis
ax.hlines(df['sizes'], xmin=df.index, xmax=-1, color='red')

# Set the x-axis ticks and labels
ax.set_xticks(df.index)
ax.set_xticklabels(df['labels'])

# Set the axis labels
ax.set_ylabel('Sizes')
ax.set_xlabel('Labels')

# Display the chart
plt.show()
```

## Donuts!

> A donut chart is similar to a pie chart, but it has a hole in the center. You can create a donut chart in Python using the matplotlib library. Hereâ€™s an example:



```python
import matplotlib.pyplot as plt

# Data to plot
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

# Create a figure and a subplot
fig, ax = plt.subplots()

# Plot the data as a pie chart
ax.pie(sizes, labels=labels)

# Add a white circle at the center to create the donut shape
circle = plt.Circle((0, 0), 0.7, color='white')
ax.add_artist(circle)

# Set the aspect ratio to 'equal' to ensure that the pie is drawn as a circle
ax.axis('equal')

# Display the chart
plt.show()
```

```python
habitat=mushrooms.groupby(['habitat']).count()
habitat
```

> Here, you are grouping your data by habitat. There are 7 listed, so use those as labels for your donut chart:



```python
labels=['Grasses','Leaves','Meadows','Paths','Urban','Waste','Wood']

plt.pie(habitat['class'], labels=labels,
        autopct='%1.1f%%', pctdistance=0.85)
  
center_circle = plt.Circle((0, 0), 0.40, fc='white')
fig = plt.gcf()

fig.gca().add_artist(center_circle)
  
plt.title('Mushroom Habitats')
  
plt.show()
```

This code draws a chart and a center circle, then adds that center circle in the chart. Edit the width of the center circle by changing 0.40 to another value.

# Waffles!

> A 'waffle' type chart is a different way to visualize quantities as a 2D array  

- We need to install a helper library called PyWaffle and use Matplotlib:

Select a segment of your data to group:



```python
capcolor=mushrooms.groupby(['cap-color']).count()
capcolor
```

Create a waffle chart by creating labels and then grouping your data:



```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle
  
data ={'color': ['brown', 'buff', 'cinnamon', 'green', 'pink', 'purple', 'red', 'white', 'yellow'],
    'amount': capcolor['class']
     }
  
df = pd.DataFrame(data)
  
fig = plt.figure(
    FigureClass = Waffle,
    rows = 100,
    values = df.amount,
    labels = list(df.color),
    figsize = (30,30),
    colors=["brown", "tan", "maroon", "green", "pink", "purple", "red", "whitesmoke", "yellow"],
)
```
